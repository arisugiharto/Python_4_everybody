```python
##searching through a file without space
fhand = open('mbox.txt')
for line in fhand:
    line=line.rstrip()
    if line.startswith('From:'):
        print(line)
```

```python
##skipping with continue
fhand = open('mbox.txt')
for line in fhand:
    line=line.rstrip()
    if not line.startswith('From:'):
        continue
    print(line)
```

```python
##Grade exercise 7.2
fname = input("Enter file name: ")
fh = open(fname)
value = 0
count = 0
for line in fh:
    if not line.startswith("X-DSPAM-Confidence:") :
        continue
    words = line.find(':')+1
    value = float(line[words:]) + value
    count = count + 1
print("Average spam confidence:",value / count)
```
